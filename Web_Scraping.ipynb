```python
# Set up the Selenium Web Driver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
wait = WebDriverWait(driver, 10)

# Open the initial page
driver.get("https://webgate.ec.europa.eu/rasff-window/screen/search")

# Find the submit button and click it
button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "button[type='submit']")))
button.click()

# Empty lists to store scraped data
categories = []
Types = []
subjects = []
dates = []
origins = []
notifies = []
Classes = []
decisions = []

# Perform actions on each page
while True:
    try:
        # Wait for the results to load
        div = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "div.results-list-body")))
        
        # Scrape data from the current page
        rows = div.find_elements(By.CSS_SELECTOR, "div.result-row")
        for row in rows:
            # Category
            cat_element = row.find_element(By.CSS_SELECTOR, "p.ng-star-inserted")
            category = cat_element.get_attribute('title') 
            
            # Type
            type_element = row.find_element(By.CSS_SELECTOR,"p.ng-star-inserted[_ngcontent-nri-c476]")
            Type = type_element.text
            
            #subjects
            subject_element = row.find_element(By.CSS_SELECTOR,"p[_ngcontent-nri-c476='']")
            subject = subject_element.text
            
            #dates
            date_element = row.find(By.CSS_SELECTOR,"p[_ngcontent-nri-c476='']")
            date = date_element.text
           
            # Origin
            origin_element = row.find(By.CSS_SELECTOR,"div[_ngcontent-nri-c476='']")
            origin = origin_element.text
            
            # Notifying
            notify_element = row.find(By.CSS_SELECTOR,"app-country[_ngcontent-nri-c476][_nghost-nri-c475]")
            notify = notify_element.text
            
            # Classes
            class_element = row.find(By.CSS_SELECTOR,"p[_ngcontent-nri-c476].ng-star-inserted")
            Class = class_element.text
            
            # Decision
            decision_element = row.find(By.CSS_SELECTOR,"p[_ngcontent-nri-c476].potential.risk.ng-star-inserted")
            decision = decision_element.text
            
            # Append data to respective lists
            categories.append(category)
            Types.append(Type)
            subjects.append(subject)
            dates.append(date)
            origins.append(origin)
            notifies.append(notify)
            Classes.append(Class)
            decisions.append(decision)
        
        # Check if there is a next page
        next_button = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "button.mat-paginator-navigation-next")))
        if "disabled" in next_button.get_attribute("class"):
            break  # Exit the loop if there is no next page

        # Perform a click action using ActionChains
        action = ActionChains(driver)
        action.move_to_element(next_button).click().perform()

        # Add a delay to allow the next page to load
        time.sleep(2)
    except Exception as e:
        print(f"An error occurred: {str(e)}")
        break

# Close the browser
driver.quit()

# Create a DataFrame from the scraped data
data = {
    "Category": categories,
    "Product": Types,
    "Subject": subjects,
    "Date": dates,
    "Origin": origins,
    "Notifying": notifies,
    "Class": Classes,
    "Decision": decisions
    
}
df = pd.DataFrame(data)

# Print the DataFrame
print(df)
```
