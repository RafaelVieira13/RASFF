__Get the RASFF Notification Reason and The Specie Affected__

Let´s use Regular Expression to find out pattern in the subject data and get the reason and the specie of each RASFF notification.

* Let´s use this site to get the regular expression:
https://regex101.com/

```python
def extract_reason_and_species(data):
    # Define the pattern using regular expressions
    pattern = r"(.*?) in (.*)"
    
    # Initialize lists to store the extracted reasons and species
    reasons = []
    species = []
    
    # Iterate over the values in the 'data' argument (assumed to be a pandas DataFrame column)
    for value in data:
        # Check if the value is a string
        if isinstance(value, str):
            # Find all matches based on the pattern in the value
            matches = re.findall(pattern, value)
            
            # Extract the reasons and species from each match and append them to the respective lists
            if matches:
                for match in matches:
                    reasons.append(match[0].strip())
                    species.append(match[1].strip())
            else:
                reasons.append(None)
                species.append(None)
        else:
            reasons.append(None)
            species.append(None)
    
    # Return the lists of reasons and species as NumPy arrays
    return np.array(reasons), np.array(species)
```

```python
reasons, species = extract_reason_and_species(bivalve['extracted_subject'])
```

```python
print("Reasons:")
for reason in reasons:
    print(reason)
```

```python
print("Species:")
for specie in species:
    print(specie)
```

```python
# Let´s create a column for Reason and Specie
bivalve['Notification Reason'], bivalve['Bivalve Specie'] = extract_reason_and_species(bivalve['extracted_subject'])
```
