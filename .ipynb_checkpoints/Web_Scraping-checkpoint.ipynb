```python
import selenium
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import time
```

__Search Element And Print The Results__

```python
service = Service(ChromeDriverManager().install())

driver = webdriver.Chrome(service=service)

driver.get("https://webgate.ec.europa.eu/rasff-window/screen/search")

# Find the element by formcontrolname attribute
wait = WebDriverWait(driver, 10)
element = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "input[formcontrolname='subject']")))

# Seacrh by element
element.send_keys("fruit")
element.send_keys(Keys.RETURN)


# Print the text from the bivalve result
try:
    div = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "div.results-list-body")))
except:
    driver.quit()
    
print(div.text)


#Print the title
#print(driver.title)

# Quit the site
driver.quit()
```

__Page Navigating and Clicking Elements__

```python
service = Service(ChromeDriverManager().install())

driver = webdriver.Chrome(service=service)

driver.get("https://webgate.ec.europa.eu/rasff-window/screen/search")

# Find the element of the button and click on the button
try:
    wait = WebDriverWait(driver, 10)
    button = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "button[type='submit']")))

    button.click()
except:
    driver.quit()
```

__ActionChains and Automating Cookie Clicker__

```python
service = Service(ChromeDriverManager().install())

driver = webdriver.Chrome(service=service)

driver.get("https://orteil.dashnet.org/cookieclicker/")

# Press the cookie as many times 
actions = ActionChains(driver)
actions.click()
actions.perform()
```
